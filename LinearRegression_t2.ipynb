```python
import pandas as pf
header=['Tanktemperature','petrol temperature','tank pressure','Petrol pressure','amount of escaping hydrocarbons']
d=pf.read_excel("C:\\Users\\Talha Habib\\Downloads\\gasdataset.xlsx",header=None,names=header)
d 
```

```python
X=d.iloc[:,0:4].values
Y=d.iloc[:,4].values
X
```

```python
from sklearn.preprocessing import Normalizer
scaler = Normalizer().fit(X)
normalizedX = scaler.transform(X)
X=normalizedX.reshape(-1,4)
X.round(3)
```

```python
import matplotlib.pyplot as ply
ply.scatter(d['Tanktemperature'],d['amount of escaping hydrocarbons'],color='red')
ply.scatter(d['petrol temperature'],d['amount of escaping hydrocarbons'],color='red')
ply.scatter(d['tank pressure'],d['amount of escaping hydrocarbons'],color='red')
ply.scatter(d['Petrol pressure'],d['amount of escaping hydrocarbons'],color='red')
ply.show()
```

```python
from sklearn.model_selection import train_test_split
x_tr,x_ts,y_tr,y_ts=train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
ab=LinearRegression()
ab.fit(x_tr,y_tr)
Y_predict=ab.predict(x_ts)
Y_predict
```

```python
from  sklearn import metrics
import numpy as n
print("Mean Absolute error",metrics.mean_absolute_error(y_ts,Y_predict))
print("Mean squared error",metrics.mean_squared_error(y_ts,Y_predict))
```
