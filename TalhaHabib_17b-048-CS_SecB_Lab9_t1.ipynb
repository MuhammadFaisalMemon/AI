```python
import pandas as pf
import numpy as n
import matplotlib.pyplot as ply
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
data=pf.read_excel('C:\\Users\\Talha Habib\\Downloads\\SalaryAge.xlsx')
x=data['YearsExperience'].values.reshape(-1,1)
y=data['Salary'].values.reshape(-1,1)
```

```python
data.plot(x='YearsExperience',y='Salary',style='o')
ply.xlabel('YearsExperience')
ply.ylabel('Salary')
ply.show()
```

```python
x_tr,x_ts,y_tr,y_ts=train_test_split(x,y,test_size=0.2)
```

```python
LR=LinearRegression()
LR.fit(x_tr,y_tr)
```

```python
y_pre=LR.predict(x_ts)
print(y_pre)
```

```python
ply.scatter(x_tr,y_tr,color='red')
ply.plot(x_tr,LR.predict(x_tr),color='blue')
ply.title("Years of experience VS Salary")
ply.xlabel('Years of experience')
ply.ylabel('Salary')
ply.show()
##here we plotted the graph on trainging dataset and fit the line in it 
```

```python
ply.scatter(x_ts,y_ts,color='blue')
ply.plot(x_ts,LR.predict(x_ts),color='red')
ply.xlabel("year of experience")
ply.ylabel("Salary")
ply.show()
## here we have plotted graph on our testing data set and fit line in it.
```

```python
print("Mean absolute error",metrics.mean_absolute_error(x_ts,y_pre))
print("Mean square error",metrics.mean_squared_error(x_ts,y_pre))
print("Root Mean square Error",n.sqrt(metrics.mean_squared_error(x_ts,y_pre)))
```
